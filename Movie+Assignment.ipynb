-  ### Subtask 3.6: Find popular genres

You might have noticed the `genres` column in the dataframe with all the genres of the movies seperated by a pipe (`|`). Out of all the movie genres, the first two are most significant for any film.

1. Extract the first two genres from the `genres` column and store them in two new columns: `genre_1` and `genre_2`. Some of the movies might have only one genre. In such cases, extract the single genre into both the columns, i.e. for such movies the `genre_2` will be the same as `genre_1`.
2. Group the dataframe using `genre_1` as the primary column and `genre_2` as the secondary column.
3. Find out the 5 most popular combo of genres by finding the mean of the gross values using the `gross` column and store them in a new dataframe named `PopGenre`.

```python
# Write your code for extracting the first two genres of each movie here
genrelist = movies.genres.str.split("|", expand = True)#extracting each genre by split operation
movies["genre_1"] = genrelist[0]#obtaining first genre and storing it
movies["genre_2"] = genrelist[1]#obtaining second genre and storing it
movies.loc[movies["genre_2"].isnull(), ["genre_2"]] = np.array(genrelist.loc[genrelist[1].isnull(), [0]] )#if second genre is null then assigning the second same as first genre
```

```python
movies_by_segment =movies.groupby(["genre_1", "genre_2"])#grouping the dataframe by genre1 and genre 2
movies_by_segment
# Write your code for grouping the dataframe here
```

```python
PopGenre =movies_by_segment.gross.mean().sort_values(ascending = False)#sorting the dataframe by finding mean of gross column
PopGenre.head(5) #printing to check the 5 most popular combo of genres
# Write your code for getting the 5 most popular combo of genres here
```

```python
# Write the code for calculating decade here
movies['decade']=movies['title_year'].apply(lambda x:x//10*10).astype(np.int32)#lambda function for calculating decade and converting to type int
movies.head()#verification for decade is added
```

```python
# Write your code for creating the data frame df_by_decade here 
df_by_decade=movies.groupby('decade')['num_voted_users'].sum().sort_values().reset_index()#Grouping decade with sum of number of voted user per decade
df_by_decade#Verification after groupby decade with sum of number of voted users
```

```python
# Write your code for plotting number of voted users vs decade
%matplotlib inline 
import matplotlib.pyplot as plt #importing matplotlib
fig,ax=plt.subplots(figsize=(10,10))#creating object of fixed size of the plot
ax.bar(df_by_decade['decade'],df_by_decade['num_voted_users'],width=7)#assigning the axis of the plot
plt.yscale('log')#Changing y-axis scale to logrithmic for better view of the bars
ax.set(title="Decade vs Number of voted users",xlabel="Decade",ylabel="Number_of_voted_users");#labelling of the plot
#As we can clearly see number of voted users after 1950 goes on increasing
```
